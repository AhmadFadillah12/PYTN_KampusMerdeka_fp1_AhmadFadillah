## E.2 Measure of Spread

```python
# Variance
variance_df = df.select_dtypes(include=[np.number]).var(ddof=1).to_frame()
variance_df.rename(columns={0: 'Variance'}, inplace=True)

# Standard Deviation
std_deviation_df = df.select_dtypes(include=[np.number]).std(ddof=1).to_frame()
std_deviation_df.rename(columns={0: 'Standard Deviation'}, inplace=True)

# Skewness
skewness_df = df.select_dtypes(include=[np.number]).skew().to_frame()
skewness_df.rename(columns={0: 'Skewness'}, inplace=True)
skewness_df['Skewness Description'] = skewness_df['Skewness'].apply(
    lambda x: 'Distribusi Positif' if x > 0 else ('Distribusi Negatif' if x < 0 else 'Distribusi Simetrik'))

# Kurtosis
kurtosis_df = df.select_dtypes(include=[np.number]).kurtosis().to_frame()
kurtosis_df.rename(columns={0: 'Kurtosis'}, inplace=True)
kurtosis_df['Kurtosis Description'] = kurtosis_df['Kurtosis'].apply(
    lambda x: 'Leptokurtik' if x > 0 else ('Platykurtik' if x < 0 else 'Mesokurtik'))

# Percentiles
percentile_values = df.select_dtypes(
    include=[np.number]).quantile([0.25, 0.50, 0.75])
percentile_values.index = ['25%', '50%', '75%']
percentile_values = percentile_values.T
percentile_values['IQR'] = percentile_values['75%'] - percentile_values['25%']

# Ranges
range_df = df.describe().loc[['min', 'max']].T
range_df.rename(columns={'min': 'Minimum', 'max': 'Maximum'}, inplace=True)

# Menggabungkan semua hasil
Measure_of_Spread = pd.concat([variance_df, std_deviation_df,
                              skewness_df, kurtosis_df, percentile_values, range_df], axis=1)

# Tampilkan hasilnya
Measure_of_Spread.reset_index().rename(columns={'index': 'Attribute'})
```

Dalam final project 1 ini kami memilih model machine learning Linear Regression sangat cocok untuk relasi yang kuat antara variabel independen dan dependen, dan cukup cepat dalam proses training model serta lebih mudah dalam di interpretasi.

Regressi Linear memiliki beberapa hyper parameter seperti :
1. fit_intercept (default = True) : Untuk Intercept 
2. positive (default = False) : Untuk prediksi harus Positif/Negatif

**Intercept (bias)** merupakan nilai ketika variabel independen (fitur) bernilai nol. Sehingga dapat mewakili harga awal sebelum dipengaruhi oleh fitur-fitur. Intercept dalam GridSearchCV diatas bernilai True dimana model kami melibatkan intercept didalam perhitungannya.

**Coef (koefisien)** merupakan nilai yang menggambarkan seberapa besar pengaruh dari fitur - fitur dalam harga. Koefisien yang bernilai positif berarti menunjukkan kenaikkan harga, namun jika bernilai negatif menunjukkan adanya kenaikan nilai dalam fitur yang terlibat dengan harga. Coef dalam hasil GridSearchCV kami diatas bernilai False (parameter “positive”) yang dapat mengindikasikan model tersebut mengizinkan koefisien berubah nilai (positif/negatif) tergantung dengan keterlibatan fitur - fitur dalam harga.

**Slope** menggambarkan sejauh mana perubahan dalam masing-masing fitur akan mempengaruhi perubahan dalam variabel target. Dalam regresi linear, kita menggunakan koefisien ini untuk memahami hubungan linear antara fitur-fitur dan variabel target. Semakin besar nilai absolut "slope" untuk suatu fitur, semakin besar pengaruhnya terhadap variabel target.

# I. Evaluation Model

**Mean Absolute Error (MAE)** merupakan cara untuk mengukur rata-rata dari selisih absolut antara y_pred (nilai prediksi) dan y_test atau y_train (nilai yang sebenarnya), dengan mengukur seberapa besar kesalahan rata-rata dari data prediksi dan data yang sebenarnya. Semakin kecil nilai MAE maka semakin baik kualitas model regresinya dan berlaku sebaliknya.

**Mean Squared Error (MSE)** merupakan cara untuk mengukur rata-rata kuadrat dari selisih antara y_pred (nilai prediksi) atau y_train/y_test(nilai yang sebenarnya), dengan memberikan nilai yang lebih besar pada kesalahan yang besar. Semakin kecil nilai MSE semakin baik juga kualitas model regresinya.

**Root Mean Squared Error (RMSE)** merupakan sebuah akar kuadrat dari MSE, yang memberikan ukuran kesalahan dalam satuan yang sama dengan variabel target (y_test/y_train). Semakin nilai RMSE kecil, juga semakin baik kualitas model regresinya.

**R-Squared (R2)** merupakan koefisien determinasi yang mengukur sejauh mana model regresi kami dapat menjelaskan variasinya dalam data. R2 bernilai 0 - 1, dimana 1 menunjukkan bahwa model dapat menjelaskan secara keseluruhan variasi dalam data, sedangkan 0 menunjukkan sebaliknya yaitu model tidak dapat menjelaskan variasi apa pun. Semakin tinggi nilai R2 maka semakin baik model dapat menjelaskan variasi di dalam data.


```python
# Evaluasi model x_test


y_pred = best_model_Linear_Regression.predict(x_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)


print("Mean Absolute Error     : ", mae)
print("Mean Squared Error      : ", mse)
print("Root Mean Squared Error : ", rmse)
print("R-squared (R2)          : ", r2)
```
